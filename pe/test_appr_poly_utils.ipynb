```python
import torch
class square(torch.nn.Module):
    def __init__(self):
        super(square, self).__init__()
    
    def forward(self, input):
        return input * input + input * input * input


a = torch.tensor(0.45, dtype=torch.float64, requires_grad=True)
print(a.grad)
output = square()
loss = output(a)
print(loss)
loss.backward()
print(a.grad)

# a_1 = a * 1
# print(a_1)
# a_2 = a * a
# print(a_2)
# a_3 = a * a * a
# print(a_3)
# a_4 = a * a * a * a
# print(a_4)

# loss = (a_1 + a_2 + a_3 + a_4)
# print(loss)
# loss.backward()

# print(a.grad)
```
